```python
import pandas as pd
from deep_translator import GoogleTranslator
import time
from tqdm import tqdm
import nltk
```

```python
df = pd.read_csv('reviews.csv', delimiter="|")
```

```python
# ‚úÖ PASO 1: Inicializar tqdm para pandas
tqdm.pandas()

def translate_reviews(df, text_column='Content'):
    """
    Traduce la columna especificada de alem√°n a ingl√©s
    """
    translator = GoogleTranslator(source='de', target='en')
    
    print(f"üîç Traduciendo {len(df)} rese√±as...")
    
    def translate_one(text):
        try:
            # Si est√° vac√≠o
            if pd.isna(text) or str(text).strip() == '':
                return ''
            
            # Peque√±a pausa para no saturar API gratuita
            time.sleep(0.15)
            
            # Traducir
            return translator.translate(str(text))
            
        except Exception as e:
            # Si falla, devolver original
            print(f"‚ö†Ô∏è Error con: {str(text)[:50]}...")
            return str(text)
    
    # ‚úÖ PASO 2: Usar .progress_apply() (ahora funciona)
    df['reviews_english'] = df[text_column].progress_apply(translate_one)
    
    print("‚úÖ Traducci√≥n completada")
    return df
```

```python
# Uso
df = translate_reviews(df, text_column='Content')
```
